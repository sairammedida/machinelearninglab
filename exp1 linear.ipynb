```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
#step 2 load the dataset
df=pd.read_csv(r"D:\insurancecharges.csv")
```

```python
#step 3 extract dependent and indcependent varible 
x=df[['bmi']]
print("independent variable values are :")
print(x)
y=df['charges']
print("dependent variables values are:")
print(y)
```

```python
#step 4 split the data into training and- testing sets
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5)
```

```python
#step 5fiting simpple linear regression to the trainning set 
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
#step 6 predicting the test set result
y_pred=model.predict(x_test)
```

```python
#evaluate the model
print(f'Mean Squared Error: {mean_squared_error(y_test,y_pred)}')
print(f'R-squared:{r2_score(y_test,y_pred)}')
```

```python
plt.scatter(x_train,y_train,color='blue',label='training data')
plt.plot(x_train,model.predict(x_train),color='red',label='regression line')
plt.xlabel=('bmi reports')
plt.ylabel=('insurance charges')
plt.title("insurance charges")
plt.legend()
plt.show()
```

```python
plt.scatter(x_test,y_test,color='green',label='Testdata')
plt.plot(x_train,model.predict(x_train),color='red',label='regression line')
plt.xlabel=("bmi reports")
plt.ylabel=('insurance charges')
plt.title("insurances charges")
plt.legend()
plt.show()
```
